```python
x_logitud = np.array([23.5, 24.5, 25.5, 26.5, 27.5, 28.5, 29.5, 30.5, 31.5,
    32.5, 33.5, 34.5, 35.5, 36.5, 37.5, 38.5, 39.5, 40.5,
    41.5, 42.5, 43.5, 44.5, 45.5, 46.5, 47.5, 48.5, 49.5], float)

y_peso = np.array([124, 146, 155, 174, 190, 213, 236, 259, 284, 308, 332, 363, 
    391, 419, 455, 500, 538, 574,
    623, 674, 724, 808, 812, 909, 1039, 1124, 1163], float)
```

#### a. Hacer un gráfico de los datos longitud versus peso de la Platija.

```python
plt.figure(figsize=(7,7), dpi=80)
plt.scatter(x_logitud,y_peso)
plt.title('Datos reales tamaño vs peso', fontsize=15)
plt.xlabel('tamaño', fontsize=15)
plt.ylabel('peso', fontsize=15)
plt.show()
```

#### b. Construir un polinomio que pase por cada uno del conjunto de datos y aproxime el peso de una platija americana cuya longitud máxima registrada es de 60cm. Concluya si es modelo construido es coherente, me representa la tendencia de la información proporcionada.

```python
pol_lagrange = ml.Lagrange(x_data=x_logitud,y_data=y_peso)
x = sp.symbols('x')
pol_evaluado = pol_lagrange.subs(x,60).evalf()

print(f"el peso aproximado para un longitud de 60cm usando el polinomio de lagrange es de P(60) = {pol_evaluado}")
```

```python
ux = np.linspace(min(x_logitud), max(x_logitud),500)
lambda_pol = sp.lambdify(x,pol_lagrange)

plt.figure(figsize=(7,7), dpi=80)
#plt.scatter(x_logitud,y_peso)
plt.plot(ux, lambda_pol(ux),color="red")
plt.title('Datos reales tamaño vs peso', fontsize=15)
plt.xlabel('tamaño', fontsize=15)
plt.ylabel('peso', fontsize=15)
plt.show()
```

## explicacion
Al construir un polinomio que pasa por los 27 puntos exactos, se genera un polinomio de grado 26. Como podrás observar en la gráfica resultante, este modelo no es coherente ni representa la tendencia general de la información. Sufre del Fenómeno de Runge, lo que significa que en los extremos del intervalo (especialmente al extrapolar a 60 cm), el polinomio oscila violentamente y arroja valores físicamente imposibles (pesos desproporcionados o incluso negativos).

### c. Construya un modelo de ajuste que represente la tendencia de los datos proporcionados y permita aproximar el peso de una platija americana cuya longitud maxima registrada es de 60 cm. Concluya si el modelo es coherente, pertinente y si su construccion es adecuada con base unicamente en los datos suministrados

```python
ml.grafiacion(x_logitud,y_peso)
```

```python
m,b= ml.minimos_cuadrados(np.log(x_logitud),np.log(y_peso))
R2 = ml.coeficiente_determinacion(np.log(x_logitud),np.log(y_peso))

modelo = lambda x : m*x + b
x_values = np.linspace(min(np.log(x_logitud)), max(np.log(x_logitud)),500)
plt.figure(figsize=(8,4), dpi=90)
plt.suptitle(" Modelo mas coherente")

plt.subplot(121)
plt.scatter(np.log(x_logitud),np.log(y_peso), color="red")
plt.plot(x_values,modelo(x_values))
plt.title(f"R²: {R2}")

plt.subplot(122)
plt.scatter(x_logitud,y_peso, color="blue")
ux = np.linspace(min(x_logitud),max(x_logitud),500)
modelo_no_lineal = lambda x: x**m * np.exp(b)
plt.plot(ux, modelo_no_lineal(ux), color="black")
plt.title("Datos originales")
```
